```python
from neupy import Pd2dBackground
import numpy
f_1 = "./full_pd2d.rcif"
with open(f_1, 'r') as fid:
    string = fid.read()
pd2d_background = Pd2dBackground()
print(pd2d_background)
```

```python
pd2d_background.from_cif(string)
print(pd2d_background)
```

```python
print(pd2d_background.to_cif)
```
